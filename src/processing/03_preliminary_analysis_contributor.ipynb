```python
from itertools import combinations
from collections import defaultdict
import csv

# Convert the start date column to datetime and sort the DataFrame chronologically
proj_hack_member["start_date_format"] = pd.to_datetime(proj_hack_member["start_date_format"])
proj_hack_member = proj_hack_member.sort_values("start_date_format").reset_index(drop=True)

# Generate a global unique participant ID mapping
all_participants = set()
proj_hack_member["contributors_list"] = proj_hack_member["contributors"].str.split(",")
for participants in proj_hack_member["contributors_list"]:
    all_participants.update(participants)
participant_id_map = {name: idx for idx, name in enumerate(all_participants)}

# Initialize a collaboration record dictionary 
# (key: sorted tuple of participant pairs, value: set of collaborated project URLs)
collaboration_map = defaultdict(set)

# Process each row to record collaborations
for _, row in proj_hack_member.iterrows():
    project_url = row["project_URL"]
    participants = [participant_id_map[name] for name in row["contributors_list"]]
    participants_sorted = sorted(participants)  # Ensure consistent keys like (a, b) instead of (b, a)
    
    # Generate all unique participant pairs in the project
    for pair in combinations(participants_sorted, 2):
        collaboration_map[pair].add(project_url)

# Write the results to a CSV file (write in chunks to avoid memory issues if needed)
with open("../data/contributor_collaboration_in_pair.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["contributor_a", "contributor_b", "projects"])
    for (a, b), projects in collaboration_map.items():
        writer.writerow([a, b, ",".join(projects)])
```

```python
# Count the number of collaborations for each participant pair
counts = [len(projects) for projects in collaboration_map.values()]

# Define buckets for grouping by collaboration frequency
bins = {
    "1 time": 0,
    "2 times": 0,
    "3–5 times": 0,
    "6 or more times": 0
}

# Assign each pair count to a bucket
for cnt in counts:
    if cnt == 1:
        bins["1 time"] += 1
    elif cnt == 2:
        bins["2 times"] += 1
    elif 3 <= cnt <= 5:
        bins["3–5 times"] += 1
    else:
        bins["6 or more times"] += 1

# Calculate total and percentage for each bucket
total_pairs = len(counts)
result = {
    "Collaboration Count": ["1 time", "2 times", "3–5 times", "6 or more times"],
    "Number of Pairs": [bins["1 time"], bins["2 times"], bins["3–5 times"], bins["6 or more times"]],
    "Percentage": [
        f"{bins['1 time']/total_pairs*100:.2f}%",
        f"{bins['2 times']/total_pairs*100:.2f}%",
        f"{bins['3–5 times']/total_pairs*100:.2f}%",
        f"{bins['6 or more times']/total_pairs*100:.2f}%"
    ]
}

# Display result as a DataFrame
result_df = pd.DataFrame(result)
print(result_df)
```

```python
# collaboration_in_pair is based on DevPost
# contributor_collaboration_in_pair is based on DevPost and GitHub
collaboration = pd.read_csv('../data/collaboration_in_pair.csv')

proj_hack_member["hackathon_list"] = proj_hack_member["hackathon_URL"].str.split(",")

project_to_hackathons = proj_hack_member.set_index("project_URL")["hackathon_list"].to_dict()

def map_projects_to_hackathons(projects_str):
    project_urls = projects_str.split(",")
    hackathon_urls = []
    for url in project_urls:
        hackathon_urls.extend(project_to_hackathons[url])
    return ",".join(hackathon_urls)

collaboration["hackathons"] = collaboration["projects"].apply(map_projects_to_hackathons)

output_columns = ["participant_a", "participant_b", "projects", "hackathons"]
collaboration[output_columns].to_csv("../data/collaboration_in_pair.csv", index=False)
```

```python
import pandas as pd
import numpy as np


split_hackathons = collaboration["hackathons"].str.split(",")

# print(split_hackathons)


first_counts = (
    split_hackathons.str[0]         
    .value_counts()
    .reset_index(name="first_devpost_colab_times")
    .rename(columns={"hackathons": "hackathon_URL"})
)

print(first_counts)

repeat_counts = (
    split_hackathons.apply(lambda x: x[1:] if len(x) > 1 else [])  
    .explode()                      
    .value_counts()
    .reset_index(name="repeat_devpost_colab_times")
    .rename(columns={"hackathons": "hackathon_URL"})
)


result_df = pd.merge(
    first_counts,
    repeat_counts,
    on="hackathon_URL",
    how="outer"
).fillna(0)

result_df[["first_devpost_colab_times", "repeat_devpost_colab_times"]] = result_df[
    ["first_devpost_colab_times", "repeat_devpost_colab_times"]
].astype(int)

total = result_df["first_devpost_colab_times"] + result_df["repeat_devpost_colab_times"]
result_df["repeat_devpost_colab_percentage"] = (
    result_df["repeat_devpost_colab_times"] / total * 100
).round(2)

bins = list(range(0, 101, 10))
labels = [f"{i}-{i+10}%" for i in range(0, 100, 10)]

result_df["percentage_bin"] = pd.cut(
    result_df["repeat_devpost_colab_percentage"],
    bins=bins,
    labels=labels,
    right=False
)

bin_counts = result_df["percentage_bin"].value_counts().sort_index().reset_index()
bin_counts.columns = ["percentage_range", "count"]


result_df.to_csv("../data/hackathon_collaboration_feature.csv", index=False)

bin_counts.to_csv("../data/percentage_distribution.csv", index=False)
```
